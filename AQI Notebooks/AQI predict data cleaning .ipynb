```python
#check missing values
df.isnull().sum()
```

```python
# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
```

```python
# Drop rows where AQI is missing
#AQI is the target variable, so rows without AQI cannot be used for training.
df = df.dropna(subset=['AQI'])
```

```python
#Fill Missing Pollutant Values
import numpy as np

# Select numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns

# Fill missing numeric values with column mean
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```

```python
# Fill AQI_Bucket with most frequent value
df['AQI_Bucket'] = df['AQI_Bucket'].fillna(df['AQI_Bucket'].mode()[0])
```

```python
# Confirm no missing values
df.isnull().sum()
```

```python
df.to_csv("cleaned_city_day.csv", index=False)
```
